```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
senti_metrics = pd.read_csv("../metrics/baseline_metrics.csv")
all_features_metrics = pd.read_csv("../metrics/best_model_metrics.csv")
```

```python
senti_metrics
```

```python
senti_mses = list(senti_metrics['MSE'])
senti_maes = list(senti_metrics['MAE'])
senti_r2s = list(senti_metrics['R2'])

all_features_mses = list(all_features_metrics['MSE'])
all_features_maes = list(all_features_metrics['MAE'])
all_features_r2s = list(all_features_metrics['R2'])
```

```python
barWidth = 0.25
fig = plt.subplots(figsize = (10,6))

br1 = [x for x in np.arange(len(senti_mses))]
br2 = [x + barWidth for x in br1]

plt.bar(br1, senti_mses, color ='red', width = barWidth, edgecolor ='black', label ='Only sentiments')
plt.bar(br2, all_features_mses, color ='green', width = barWidth, edgecolor ='black', label ='All features')

plt.xlabel('Model', fontweight ='bold', fontsize = 15)
plt.ylabel('MSE', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(senti_mses))], ['Linear Regression', 'Lasso Regression', 'Ridge Regression', 'SGDRegressor', 'SVR'])

plt.legend()
plt.savefig("../images/models_mse_figure.png")
plt.show()
```

```python
barWidth = 0.25
fig = plt.subplots(figsize = (10,6))

br1 = [x for x in np.arange(len(senti_maes))]
br2 = [x + barWidth for x in br1]

plt.bar(br1, senti_maes, color ='red', width = barWidth, edgecolor ='black', label ='Only sentiments')
plt.bar(br2, all_features_maes, color ='green', width = barWidth, edgecolor ='black', label ='All features')

plt.xlabel('Model', fontweight ='bold', fontsize = 15)
plt.ylabel('MAE', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(senti_maes))], ['Linear Regression', 'Lasso Regression', 'Ridge Regression', 'SGDRegressor', 'SVR'])

plt.legend()
plt.savefig("../images/models_mae_figure.png")
plt.show()
```
